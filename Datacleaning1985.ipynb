```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df=pd.read_csv('auto.csv',names=headers)
```

```python
df.head()
```

```python
##  there are many ? in dataframe 
## we need to identify all those values
```

```python
## first we need to convert "?" to NaN(Not a Number)
```

```python
df.replace('?',np.NaN,inplace=True)
```

```python
df.head()
```

```python
## number of missing values

missing_data=df.isnull().sum()
missing_data.sort_values(inplace=True,ascending=False)
missing_data.head(8)
```

```python
# claculate the avg of the columns
```

```python
avg_norm_loss=df['normalized-losses'].astype('float').mean(axis=0)
print('avg of normalized-losses',avg_norm_loss)
avg_bore=df['bore'].astype('float').mean(axis=0)
print('bore',avg_bore)
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print('stroke',avg_stroke)
avg_horsepower=df['horsepower'].astype('float').mean(axis=0)
print('horsepower',avg_horsepower)
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print('peak-rpm',avg_peakrpm)
```

```python
#Replace "NaN" by mean value in columns
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
df["bore"].replace(np.nan, avg_bore, inplace=True)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
df.dropna(subset=['price'],axis=0,inplace=True)#drop whole row with NaN in "price" column
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df.dtypes
```

```python
## in the data set the bore and stroke column should be int or float 
## but they are identified as object suggesting it isn't completely numerical  
```

```python
df[['bore','stroke']]= df[['bore','stroke']].astype('float')
df[['normalized-losses']]= df[['normalized-losses']].astype('int')
df[['peak-rpm']]= df[['peak-rpm']].astype('float')
df[['price']]= df[['price']].astype('float')
```

```python
df.dtypes
```

```python
#transform mpg to L/100km  ##  L/km is more commonly used
    
```

```python
###  THE FORMULA ##
#  L/100Km=235/mpg
```

```python
df.head()
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km']=235/df['city-mpg']
```

```python
df.head() ## we can see at the end there is an additional column forcity-L/100km
```

```python
# similarly we can transform mpg to L/100km by mathematical operation (235 divided by mpg)
```

```python
df['highway-L/100Km']=235/df['highway-mpg']
```
